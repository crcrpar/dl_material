```python
%matplotlib inline
import numpy as np
import scipy
import matplotlib.pyplot as plt

def gen_random(loc, size):
    return list(np.random.normal(loc=loc, scale=0.25, size=size))

# fig 1
x0, x5, x00, x55 = gen_random(0.0, 100), gen_random(5.0, 100), gen_random(0.0, 100), gen_random(5.0, 100)
a_x = x0 + x5
a_y = x00 + x55
b_x = x5 + x0
b_y = x0 + x5
x= np.linspace(0.0, 5.0, 100)
y = x - 0.5
group_a = plt.scatter(a_x, a_y, alpha=0.6)
group_b = plt.scatter(b_x, b_y, alpha=0.6)
plt.plot(x, y, 'g', label='bound')
plt.legend((group_a, group_b), ('group A', 'group B'), bbox_to_anchor=(1.3, 1))
plt.legend()
plt.title('fig 1. problem ')
plt.show()
plt.savefig('./images/fig_01.png')
plt.close('all')

# fig 2
a_x = x0 + x5
a_y = x00 + x0
b_x = x5 + x0
b_y = x55 + x5
x = [2.5] * 100
y = np.linspace(-0.2, 5.2, 100)
group_a = plt.scatter(a_x, a_y, alpha=0.6)
group_b = plt.scatter(b_x, b_y, alpha=0.6)
plt.plot(x, y, 'g', label='bound')
plt.legend((group_a, group_b), ('group A', 'group B'), bbox_to_anchor=(1.3, 1))
plt.legend()
plt.title('fig 2. linear separable problem')
plt.show()
plt.savefig('./images/fig_02.png')
plt.close('all')

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def step(x):
    tmp = x > 0
    return tmp.astype(int)

def relu(x):
    return np.minimum(np.maximum(0.25 * x, 0), 6)

x = np.linspace(-5, 5, 1000)
y1 = sigmoid(x)
y2 = step(x)
y3 = relu(x)
y4 = np.tanh(x)
y5 = x * x / 25.0
plt.plot(x, y2, label='step')
plt.title('fig 3. Step')
plt.legend(bbox_to_anchor=(1.3, 1))
plt.show()
plt.savefig('./images/fig_03.png')
plt.close('all')
plt.plot(x, y1, label='sigmoid')
plt.xlabel('x')
plt.ylabel('y')
plt.title('fig 4. Sigmoid')
plt.legend(bbox_to_anchor=(1.3, 1))
plt.show()
plt.savefig('./images/fig_04.png')
plt.close('all')
plt.plot(x, y3, label='relu')
plt.xlabel('x')
plt.ylabel('y')
plt.title('fig 5. ReLU')
plt.legend(bbox_to_anchor=(1.3, 1))
plt.show()
plt.savefig('./images/fig_05.png')
plt.close('all')
plt.plot(x, y4, label='tanh')
plt.xlabel('x')
plt.ylabel('y')
plt.title('fig 6. tanh')
plt.legend(bbox_to_anchor=(1.3, 1))
plt.show()
plt.savefig('./images/fig_06.png')
plt.close('all')
plt.plot(x, y5, label='squared error')
plt.xlabel('x')
plt.ylabel('y')
plt.title('fig 7. squared error')
plt.legend(bbox_to_anchor=(1.4, 1))
plt.show()
plt.savefig('./images/fig_07.png')
plt.close('all')
```
